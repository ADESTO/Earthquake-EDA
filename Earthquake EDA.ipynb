```python
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt

# Create a GeoDataFrame from your DataFrame
# This assumes that your DataFrame is named df and has columns 'Longitude' and 'Latitude'
geometry = [Point(xy) for xy in zip(data['Longitude'], data['Latitude'])]
gdf = gpd.GeoDataFrame(data, geometry=geometry)

# Create a world map
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Plot the world map
fig, ax = plt.subplots(figsize=(15, 10))
world.plot(ax=ax, color='white', edgecolor='black')

# Plot the earthquake data
gdf.plot(ax=ax, color='red', markersize=10)

plt.show()

```

```python
# Create a GeoDataFrame from your DataFrame
geometry = [Point(xy) for xy in zip(data['Longitude'], data['Latitude'])]
gdf = gpd.GeoDataFrame(data, geometry=geometry)

# Load a GeoDataFrame with country boundaries
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Filter for African countries
africa = world[world['continent'] == 'Africa']

# Ensure both GeoDataFrames are in the same CRS
gdf.crs = "EPSG:4326"
africa = africa.to_crs(gdf.crs)

# Use a spatial join to find earthquakes in Africa
african_earthquakes = gpd.sjoin(gdf, africa, op='within')

# Now `african_earthquakes` is a DataFrame containing only the earthquakes that occurred in Africa
# Plot the world map
fig, ax = plt.subplots(figsize=(12, 5))
africa.boundary.plot(ax=ax, color='black')
african_earthquakes.plot(ax=ax, color='white', edgecolor='black')

# Plot the earthquake data
african_earthquakes.plot(ax=ax, color='blue', markersize=10)

plt.show()
```

```python
# Find the number of earthquakes
num_earthquakes = len(african_earthquakes)

# Find the unique locations
unique_locations = african_earthquakes['name'].unique()

# Group by 'name' and 'Year', then find unique values
unique_locations_by_year = african_earthquakes.groupby(['name', 'Year']).size().reset_index().rename(columns={0:'count'})

print(unique_locations_by_year)


print(f"Number of earthquakes in Africa: {num_earthquakes}")
print(f"Unique locations of earthquakes in Africa: {unique_locations}")
```

```python
# Calculate the country with the most earthquakes
most_earthquakes_country = african_earthquakes['name'].value_counts().idxmax()

print(f"The country in Africa that experienced the most earthquakes is: {most_earthquakes_country}")
```

```python
# Calculate the number of earthquakes for the country
country = most_earthquakes_country
num_earthquakes = african_earthquakes[african_earthquakes['name'] == country].shape[0]

print(f"The country {country} experienced {num_earthquakes} earthquakes according to the dataset, making it the most affected by earthquakes.")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
import pandas as pd

data = pd.read_csv("C:\\Users\Ade\Downloads\\significant earthquake\Significant Earthquake Dataset 1900-2023.csv")
# Assume you have a DataFrame `df` with your earthquake data

# Select features and target
features = data[['Latitude', 'Longitude', 'Depth']]
target = data['Mag']

# Split the data into training and test sets
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Create a decision tree regressor
model = DecisionTreeRegressor(random_state=42)

# Train the model
model.fit(features_train, target_train)

# Use the model to make predictions
predictions = model.predict(features_test)

```

```python
print(predictions)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(target_test, predictions)
print(f"Mean Squared Error: {mse}")
```

```python
# Create a DataFrame with the actual values and predictions
results = pd.DataFrame({'Actual': target_test, 'Predicted': predictions})

# Reset the index of the DataFrame
results.reset_index(drop=True, inplace=True)

# Print the first 15 rows of the DataFrame
print(results.head(15))
```
